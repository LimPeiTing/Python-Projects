```python
# Gender
sns.countplot(data = df, x = "Gender", hue = "Gender", palette = {"M":"#0071ce", "F":"#ffc220"}, legend = True )
plt.title("Gender VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("Gender")

plt.show()
```

```python
# From the plot, it can be observed that Male make the most of the purchase. 
```

```python
# We will now look into Marital_Status. In the dataset, 0 indicates unmarried, 1 indicates married. 
# We will rename 0 and 1 to unmarried and married. But before that, we will create a copy of the dataframe to modify on this, to ensure that the original dataframe would not be affected

# Marital Status 

# Create a copy of the dataset 
df2 = df.copy()
df2.loc[df2['Marital_Status'] == 0, 'Marital_Status'] = "Unmarried"
df2.loc[df2['Marital_Status'] == 1, 'Marital_Status'] = "Married"

# Plot 
sns.countplot(data = df2, x = "Marital_Status", palette = {"Unmarried":"#0071ce", "Married":"#ffc220"}, hue = "Marital_Status", legend = True)
plt.title("Marital Status VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("Marital Status")

plt.show()
```

```python
# List of Common Seaborn Palettes
# Categorical: "deep", "muted", "bright", "pastel", "dark", "colorblind"
# Sequential: "Blues", "BuGn", "BuPu", "GnBu", "Greens", "Greys", "Oranges", "Purples", "Reds"
# Diverging: "coolwarm", "seismic", "Spectral"
# Each of these palettes can be accessed with sns.color_palette("palette_name", n_colors), where n_colors is the number of colors you want.
```

```python
# Age Distribution

desired_hue_order = ['0-17', '18-25', '26-35', '36-45','46-50', '51-55', '55+']

sns.countplot(data = df2, x = "Age", palette = "bright", order= sorted(df2['Age'].unique()), hue= "Age", hue_order= desired_hue_order, legend= True)
plt.title("Age Group VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("Age Group")
plt.show()

# .unique: to retrieve all unique values 
# .sorted: to sort the x-axis data in order 
```

```python
# From above plot, it can be observed that majority of Walmart's user are coming from age group of 26-35, followed by 36-45 and 18-25. 
```

```python
# City Purchase

desired_hue_order = ['A', 'B', 'C']

sns.countplot(data = df2, x = "City_Category",  palette = "dark", order = sorted(df2['City_Category'].unique()), hue= "City_Category", hue_order= desired_hue_order, legend= True)
plt.title("City Category VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("City Category")
plt.show()
```

```python
# Above plot showed that City B has more customer compared with City A and C. 
```

```python
# .size() calculates the size of each group, i.e., the number of occurrences of each unique "Product_Category"
product_df = df.groupby(['Product_Category']).size()
product_df
```

```python
# .to_frame(): Converts the resulting Series into a DataFrame, naming the new column "count_of_occupation"
# This results in a DataFrame where "Product_Category" is the index, and "count_of_occupation" contains the count of occurrences

product_df = df.groupby(['Product_Category']).size().to_frame("count_of_product")
product_df
```

```python
# .reset_index: Resets the index of the DataFrame so that "Product_Category" becomes a regular column rather than the index
product_df.reset_index()
```

```python
product_df = df.groupby(['Product_Category']).size().to_frame("count_of_product").reset_index()
product_df
```

```python
# City Category Purchase

product_df = df.groupby(['Product_Category']).size().to_frame("count_of_product").reset_index()

top_3_product_count = product_df['count_of_product'].nlargest(3).values

colors = ["#ffc220" if x in top_3_product_count else "gray" for x in product_df['count_of_product']]

sns.barplot(data= product_df, x= "Product_Category", y="count_of_product", palette=colors, hue= "Product_Category", legend = False)
plt.title("Product Category VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("Product Category")
plt.show()
```

```python
# It can be concluded that Product Category 1,5 and 8 are the top 3 products with the highest purchase count. 
```

```python
# Occupation Purchase

occupation_df = df.groupby(['Occupation']).size().to_frame("count_of_occupation").reset_index()

top_3_occ_count = occupation_df['count_of_occupation'].nlargest(3).values

colors = ["#ffc220" if x in top_3_occ_count else "gray" for x in occupation_df['count_of_occupation']]

sns.barplot(data= occupation_df, x= "Occupation", y="count_of_occupation", palette=colors, hue= "Occupation", legend = False)
plt.title("Occupation VS Count of Purchase")
plt.ylabel("Count of Purchase")
plt.xlabel("Occupation")
plt.show()
```

```python
# It can be concluded that Occupation 0, 4 and 7 are the top 3 products with the highest purchase count. 
```

```python
# Distribution of Purchase Amount
plt.figure(figsize= (10, 6))
sns.histplot(df['Purchase'], bins= 30, kde= True)
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# KDE estimates the underlying distribution of the data by using a kernel (often a Gaussian, or "bell-shaped," curve) for each data point.
# The result is a smooth curve that approximates the distribution of the data, making it easier to see the general pattern without the blocky appearance of the histogram bars.

# In matplotlib, plt.tight_layout() is used to automatically adjust the spacing between subplots, ensuring that they fit within the figure area without overlapping. This function is particularly helpful when labels, titles, or tick marks are being cut off or when multiple subplots are tightly packed together, as it reduces excess whitespace and makes the plot more readable.
```

```python
# Top 10 Product

# 1. Count occurrences of each Product_ID

product_counts = df['Product_ID'].value_counts()
```

```python
# 2. Select the top 10 most frequent Product_IDs

top_10_products = product_counts.head(10)
```

```python
# 3. Plot the bar graph

plt.figure(figsize=(10, 6))
sns.barplot(x= top_10_products.index, y= top_10_products.values, palette= "viridis", hue= top_10_products.index, legend= False)
plt.title("Top 10 Most Frequent Product IDs")
plt.xlabel("Product ID")
plt.ylabel("Frequency")
plt.xticks(rotation=45)  # Rotate x labels if needed for better readability
plt.tight_layout()
plt.show()
```
